#import pandas

```python
import pandas as pd
```

#Read in the dataset

```python
hotel_bookings = pd.read_csv('hotel_bookings.csv')
```

#First look at the dataset

```python
hotel_bookings
```

#Drop/rename columns

```python
hotel_bookings.columns
```

```python
hotel_bookings.drop(columns=['Unnamed: 0'], inplace=True, errors='ignore')
hotel_bookings.rename(columns={'adults': 'num_adults', 'children': 'num_children', 'babies': 'num_babies'}, inplace=True)
hotel_bookings.head(10)
```

#NaNs

```python
hotel_bookings.isnull().sum() *100 / len(hotel_bookings)
```

```python
hotel_bookings['agent'].unique()
hotel_bookings[hotel_bookings['agent'] == 5]

hotel_bookings['agent'] = hotel_bookings['agent'].fillna(-1).astype(int)

hotel_bookings[hotel_bookings['num_children'].isna()]

hotel_bookings[hotel_bookings['country'].isna()][['hotel', 'is_canceled', 'country']]

hotel_bookings['country'] = hotel_bookings['country'].fillna('Unknown')

hotel_bookings.dropna(subset=['num_children'], inplace=True)
```

#remove the 'company' column from the dataset here

```python
hotel_bookings.drop(columns=['company', 'Unnamed: 0'], inplace=True, errors='ignore')
```

#Check column data types

```python
hotel_bookings.dtypes
```

#Change data type

```python
hotel_bookings = hotel_bookings.astype({'is_canceled': 'boolean', 'is_repeated_guest': 'boolean', 'num_children': 'int64', 'agent': 'int'})
hotel_bookings.dtypes
```

#Bin columns

```python
hotel_bookings['lead_time'].unique
hotel_bookings['lead_time'].describe()

#define the bins
bins = [0, 100, 200, 300, 400, 500, 600, 700, 800]
labels = ['0-100', '101-200', '201-300', '301-400', '401-500', '501-600', '601-700', '701-800'] 

hotel_bookings['lead_time_binned'] = pd.cut(hotel_bookings['lead_time'], bins=bins, labels=labels)
hotel_bookings[['lead_time', 'lead_time_binned']]
```

#Separate Columns

```python
hotel_bookings.head()
```

```python
hotel_bookings['arrival_date'] = (hotel_bookings['arrival_date_year'].astype(str) 
                                  + '-' + hotel_bookings['arrival_date_month'] 
                                  + '-' + hotel_bookings['arrival_date_day_of_month'].astype(str))
hotel_bookings.head()

column_to_move = hotel_bookings.pop('arrival_date_month')
hotel_bookings.insert(4, 'arrival_date_month', column_to_move)
hotel_bookings.head()

column_to_move = hotel_bookings.pop('arrival_date_year')
hotel_bookings.insert(5, 'arrival_date_year', column_to_move)
hotel_bookings.head()

 
```

```python
#identify unique values in 'hotel' column(typos/errors)
hotel_bookings['hotel'].unique()
```

#String cleaning

```python
hotel_bookings['hotel'] = hotel_bookings['hotel'].replace(f"[\*\n\^]", '', regex=True)
hotel_bookings['hotel'].unique()
```

#Remove duplicates in the dataframe, focus on cases where every column value is duplicated.

```python
hotel_bookings.loc[hotel_bookings.duplicated(keep=False)]
hotel_bookings.drop_duplicates(keep='first', inplace=True)
```
